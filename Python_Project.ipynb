```python
# TASK 1

# Determined the distribution of employees across each team 
# and calculated the percentage split relative to the total number of employees

import pandas as pd
import numpy as np

# Preprocessing

data =open(r"C:/Users/HP/Downloads/myexcel - myexcel.csv.csv")
data = pd.read_csv(r"C:/Users/HP/Downloads/myexcel - myexcel.csv.csv")


# Replaced "height" column values with random numbers
data['height'] = np.random.randint(150, 181, size=len(data))

# Ensured the data consistency and integrity
print(data.info())
print(data.head())
```

```python
# TASK 2

team_distribution = data['Team'].value_counts(normalize=True) * 100
print(team_distribution)
```

```python
import matplotlib.pyplot as plt

team_distribution.plot(kind='bar', title="Employee Distribution by Team", color='skyblue')
plt.ylabel("Percentage")
plt.show()
```

```python
# TASK 3

# Segregated employees based on their positions within the company.

position_distribution = data['Position'].value_counts()
print(position_distribution)
```

```python
import matplotlib.pyplot as plt

position_distribution = data['Position'].value_counts()

position_distribution.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Distribution of Employees by Position")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.show()
```

```python
# TASK 4

# Identified the predominant age group among employees.

bins = [20, 30, 40, 50, 60]
labels = ['20-30', '31-40', '41-50', '51-60']
data['Age_group'] = pd.cut(data['Age'], bins=bins, labels=labels)

age_group_distribution = data['Age_group'].value_counts()
print(age_group_distribution)

```

```python
import matplotlib.pyplot as plt

plt.hist(data['Age'], bins=[20, 30, 40, 50, 60], edgecolor='black', color='skyblue')
plt.title("Age Distribution of Employees")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.xticks([25, 35, 45, 55], labels=['20-30', '31-40', '41-50', '51-60'])

plt.show()
```

```python
# TASK 5

# Discovered the team and its position which have the highest salary expenditure.

salary_team = data.groupby('Team')['Salary'].sum()
salary_position = data.groupby('Position')['Salary'].sum()

print("Team with highest salary is:", salary_team.idxmax())
print("Position with highest salary is:", salary_position.idxmax())
```

```python
# Investigated the correlation between age and salary and represented it visually.

correlation = data['Age'].corr(data['Salary'])
print("Correlation between Age and Salary:", correlation)

plt.scatter(data['Age'], data['Salary'], alpha=0.5, color='green')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
